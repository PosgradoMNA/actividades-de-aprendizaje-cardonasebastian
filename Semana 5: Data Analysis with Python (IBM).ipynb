```python
#Importar librerias.
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
#declarando variable con la url de nuestro archivo online.
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
#colocando nuestros headers para poder cambiarlos por los default que vienen en nuestro dataset.
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#leemos nuestro archivo con la funcion read_csv con la libreria de pandas.
df = pd.read_csv(filename, names = headers)
```

```python
# Imprimimos las primeras 5 lineas de nuestro dataframe
df.head()
```

```python
#Calcular el valor medio de la columna "wheel-base"
avg_norm_loss = df["wheel-base"].astype("float").mean(axis=0)
print("Average of wheel-base:", avg_norm_loss)
```

```python
#Calcular el valor medio de la columna "bore"
avg_bore=df['bore'].head().astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
df['wheel-base'].value_counts()
```

```python
df['wheel-base'].value_counts().idxmax()
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Convertir mpg a L/100km por operación matemática (235 dividido por mpg)
df['city-L/100km'] = 235/df["city-mpg"]

# revisar el dato transformado.
df.head()
```

```python
#reemplazar (valor original) por (valor original)/(valor máximo)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
# mostrar las columnas escaladas
df[["length","width","height"]].head()
```

```python
df["wheel-base"]=df["wheel-base"].astype(int, copy=True)
df
```

```python
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["wheel-base"])

# asignando x/y etiquetas y el titulo
plt.pyplot.xlabel("wheel-base")
plt.pyplot.ylabel("count")
plt.pyplot.title("wheel-base bins")
```

```python
bins = np.linspace(min(df["wheel-base"]), max(df["wheel-base"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
df['wheel-base-binned'] = pd.cut(df['wheel-base'], bins, labels=group_names, include_lowest=True )
df[['wheel-base','wheel-base-binned']].head(20)
```

```python
df["wheel-base-binned"].value_counts()
```

```python

pyplot.bar(group_names, df["wheel-base-binned"].value_counts())

# asignando x/y etiquetas y el titulo
plt.pyplot.xlabel("wheel-base")
plt.pyplot.ylabel("count")
plt.pyplot.title("wheel-base bins")
```

```python


# dibuja el histograma del atributo "wheel-base" con contenedores = 3
plt.pyplot.hist(df["wheel-base"], bins = 3)

# asignando x/y etiquetas y el titulo
plt.pyplot.xlabel("wheel-base")
plt.pyplot.ylabel("count")
plt.pyplot.title("wheel-base bins")
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# combinar el dataframe "df" y "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)

# eliminar la columna original "tipo de combustible" de "df"
df.drop("fuel-type", axis = 1, inplace=True)

#Imprimir dataframe
df
```
