```python
#Importar librerias.
import pandas as pd
import matplotlib.pylab as plt
```

```python
#declarando variable con la url de nuestro archivo online.
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
#colocando nuestros headers para poder cambiarlos por los default que vienen en nuestro dataset.
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#leemos nuestro archivo con la funcion read_csv con la libreria de pandas.
df = pd.read_csv(filename, names = headers)
```

##Identidicar valores faltantes

```python
#Importando libreria para poder realizar operaciones numericas de entradas y salidas con python.
import numpy as np
```

```python
# remplazar el simbolo "?" con NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

##Evalucacion de datos faltantes.

```python
# verificamos que datos estan nulos y nos retona True o False dependindo del dato
missing_data = df.isnull()
missing_data.head(5)
```

##Contando los valores faltantes de cada columna.

```python
#Usamos un bucle for en Python, se puede calcular rápidamente la cantidad de valores que faltan en cada columna.
# En el cuerpo del bucle for, el método ".value_counts()" cuenta el número de valores "Verdaderos".
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")

#Basado el resumen anterior, cada columna tiene 205 filas de datos y siete de las columnas contienen datos faltantes.
```
